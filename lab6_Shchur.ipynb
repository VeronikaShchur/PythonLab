## Завдання 1. Створити програму для генерації довільного масиву із 42 цілих чисел, які не перевищують по модулю 99, з наступними властивостями:

1.1. При кожному запуску програми масив повинен бути різним\
1.2. Масив повинен містити 40 відсотків від’ємних елементів\
1.3. Масив повинен містити 10 відсотків нульових елементів (розташованих в довільних місцях)

```python
import random

n = 42

# Відсоток від'ємних елементів (40% від 42)
negative_count = int(n * 0.40)

# Відсоток нульових елементів (10% від 42)
zero_count = int(n * 0.10)

# Кількість інших випадкових елементів (решта)
other_count = n - negative_count - zero_count

# Генеруємо 40% від'ємних елементів у діапазоні [-99, -1]
negative_elements = [random.randint(-99, -1) for _ in range(negative_count)]

# Генеруємо 10% нульових елементів
zero_elements = [0] * zero_count

# Генеруємо решту елементів в діапазоні [1, 99]
other_elements = [random.randint(1, 99) for _ in range(other_count)]

# Об'єднуємо всі елементи в один масив
result_array = negative_elements + zero_elements + other_elements

# Перемішуємо масив для випадкового розташування елементів
random.shuffle(result_array)

# Виводимо результат
print(result_array)
```

## Завдання 2. Зберегти згенерований масив в файл

```python
# Запис масиву у файл
with open('array.txt', 'w') as file:
    file.write(' '.join(map(str, result_array)))

print("Масив збережено у файл 'array.txt'!")
```

## Завдання 3. Створити програму, яка зчитує із файлу згенерований масив попередньою програмою і виводить його на екран

```python
# Зчитування масиву з файлу
with open('array.txt', 'r') as file:
    # Зчитуємо вміст файлу
    content = file.read()

# Перетворюємо рядок у список цілих чисел
array_from_file = list(map(int, content.split()))
print(array_from_file)
```

## Завдання 4. Візуалізувати масив за допомогою стовпчастої горизонтальної діаграми. Виділити кольором мінімальний елемент, а також вказати значення на діаграмі. Виділити іншим кольором від’ємні елементи

```python
import matplotlib.pyplot as plt

# Діаграма для візуалізації масиву
min_value = min(array_from_file)  # Мінімальне значення масиву

# Створення кольорів для кожного елементу масиву
colors = ['red' if x == min_value else 'blue' if x < 0 else 'green' for x in array_from_file]

# Створюємо горизонтальну стовпчасту діаграму
plt.barh(range(len(array_from_file)), array_from_file, color=colors)

# Додаємо значення на діаграму
for index, value in enumerate(array_from_file):
    plt.text(value, index, str(value), va='center', ha='left', color='white' if value < 0 else 'black')

# Оформлення діаграми
plt.xlabel('Значення')
plt.ylabel('Індекс елемента')
plt.title('Візуалізація масиву')

# Показуємо діаграму
plt.show()
```

## Завдання 5. Порахувати суму від’ємних елементів

```python
# Обчислення суми від'ємних елементів
negative_sum = sum(x for x in array_from_file if x < 0)

# Виведення результату
print(f"Сума від'ємних елементів: {negative_sum}")
```

## Завдання 6. Порахувати кількість від’ємних та кількість додатних елементів масиву, розташованих після мінімального

```python
# Знаходимо індекс мінімального елемента в масиві
min_element_index = array_from_file.index(min(array_from_file))

# Підмасив, який містить елементи після мінімального
elements_after_min = array_from_file[min_element_index + 1:]

# Підрахунок кількості від'ємних елементів після мінімального
negative_count_after_min = sum(1 for x in elements_after_min if x < 0)

# Підрахунок кількості додатних елементів після мінімального
positive_count_after_min = sum(1 for x in elements_after_min if x > 0)

# Виведення результатів
print(f"Кількість від'ємних елементів після мінімального: {negative_count_after_min}")
print(f"Кількість додатних елементів після мінімального: {positive_count_after_min}")
```

## 7. Замінити в масиві всі від’ємні елементи їх квадратами. Відсортувати масив за зростанням. Вивести отриманий масив та його графічну діаграму на екран

```python
# Замінюємо всі від'ємні елементи їх квадратами
squared_array = [x**2 if x < 0 else x for x in array_from_file]

# Сортуємо масив за зростанням
sorted_array = sorted(squared_array)

# Виводимо результат
print("Отриманий масив після заміни від'ємних елементів їх квадратами та сортування:")
print(sorted_array)

# Побудова графічної діаграми
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_array)), sorted_array, color='skyblue')

# Додавання підписів та заголовку
plt.xlabel('Значення')
plt.ylabel('Індекс елемента')
plt.title('Графічне представлення модифікованого масиву')

# Відображення значень на діаграмі
for i, value in enumerate(sorted_array):
    plt.text(value, i, str(value), va='center', ha='left', color='black')

# Показуємо графік
plt.show()
```
